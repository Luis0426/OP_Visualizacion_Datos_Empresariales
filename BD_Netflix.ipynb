```python
#DATA FRAME
dataset = pd.read_csv('https://raw.githubusercontent.com/Luis0426/OP_Visualizacion_Datos_Empresariales/refs/heads/main/NetFlix.csv', sep = ',' )

dataset.head()

```

# **VER TIPOS**

```python
dataset.dtypes
```

```python
print(dataset['date_added'].head())
```

```python
#     convertir directamente, capturando errores
try:
    dataset['date_added'] = pd.to_datetime(dataset['date_added'])
except ValueError:
    dataset['date_added'] = pd.to_datetime(dataset['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
#verificar el tipo de datos despues de la conversion
print(dataset['date_added'].head())
```

```python
print(dataset['date_added'].head())


#Si hay valores NaT (Not a Time) despues de la conversion los podemos identificar asi
print(dataset['date_added'].isna().sum())
```

```python

rating_map = {
    'G': 1,
    'PG': 2,
    'PG-13': 3,
    'R': 4,
    'NC-17': 5,
    'TV-Y': 1,
    'TV-Y7': 2,
    'TV-G': 1,
    'TV-PG': 2,
    'TV-14': 3,
    'TV-MA': 4
}
```

```python
#cREAR UNA NUEVA COLUMNA 'RATING NUMERIC' CON LOS VALORES CONVERTIDOS
dataset['rating_numeric'] = dataset['rating'].map(rating_map)
```

```python

#Mostrar las primeras filas para verificar la conversion
print(dataset[['ratiing', 'rating_numeric']].head())

#Obtener estadisticas basicas de la nueva columna numerica
print(dataset['rating_numeric'.describe])
```
