# IMPORTAR LIBRERIAS

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos de Ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

# Grafico de Lineas

```python
#Grafico de lineas
plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.title('Gráfico de Línea')
plt.xlabel('Eje X')
plt.ylabel('sin(x)')
plt.show()
```

# GRAFICO DE BARRAS

```python
categorias = ['A', 'B', 'C', 'D']
valores = [3, 9, 1, 5]

plt.figure(figsize=(10, 4))
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```

# GRAFICO DE DISPERSION

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 4))
plt.scatter(x, y)
plt.title('Gráfico de Diseprsion')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# HISTOGRAMA

```python
datos = np.random.randn(1000)

plt.figure(figsize=(10, 4))
plt.hist(datos, bins=30)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
